```python
# ETS Decompositions
from statsmodels.tsa.seasonal import seasonal_decompose
# Q1
# Using ETS Decomposition displaying the seasonality alone (classifie)d_ts_06.csv), what can you tell about the presence of seasonality?
# Support your answer with a graph of size (30, 5)

# Perform ETS decomposition (Additive Model)
result = seasonal_decompose(classified_ts_06.loc['1980':'1984', 'Value'], model='additive')

# Plot seasonality component
plt.figure(figsize=(30,5))
result.seasonal.plot(marker='o')

# Title and grid
plt.title("ETS Seasonality (TS6) [1980-1984]")
plt.grid()
plt.show()
```

```python
# Q2
# Using ETS Decomposition displaying the trend graph alone (classified_ts_06.csv), what is the type of trend?
# Support your answer with a graph of size (30, 5)

result = seasonal_decompose(classified_ts_06['Value'], model='additive')
result.trend.plot(figsize = (30,5))
plt.title("ETS Trend (TS6)")
plt.grid()
plt.tight_layout()
plt.show()
```

```python
  # Q12
  # Generate the graphs below with atleast 90% similarity ('On Notebook')
  # Subplot size is (35, 20)

fig, axes = plt.subplots(4, 4, figsize=(35, 20))

# ETS Decompositions
ets_classified_ts_06 = seasonal_decompose(classified_ts_06['Value'], model='additive')
ets_classified_ts_07 = seasonal_decompose(classified_ts_07['Value'], model='additive')

# STL Decompositions (using period=4 and robust=True)
stl_classified_ts_06 = STL(classified_ts_06["Value"], period=4, robust=True).fit()
stl_classified_ts_07 = STL(classified_ts_07["Value"], period=4, robust=True).fit()

titles = [
    ["Original Series (TS6)", "ETS Trend (TS6)", "ETS Seasonality (TS6)", "ETS Residuals (TS6)"],
    ["Original Series (TS6)", "STL Trend (TS6)", "STL Seasonality (TS6)", "STL Residuals (TS6)"],
    ["Original Series (TS7)", "ETS Trend (TS7)", "ETS Seasonality (TS7)", "ETS Residuals (TS7)"],
    ["Original Series (TS7)", "STL Trend (TS7)", "STL Seasonality (TS7)", "STL Residuals (TS7)"]
]

# TS6 ETS components (Column 0)
components_ets_ts6 = [
    classified_ts_06["Value"],
    ets_classified_ts_06.trend,
    ets_classified_ts_06.seasonal,
    ets_classified_ts_06.resid
]
for i, component in enumerate(components_ets_ts6):
    component.plot(ax=axes[i, 0], title=titles[0][i])
    axes[i, 0].grid()

# TS6 STL components (Column 1)
components_stl_ts6 = [
    classified_ts_06["Value"],
    stl_classified_ts_06.trend,
    stl_classified_ts_06.seasonal,
    stl_classified_ts_06.resid
]
for i, component in enumerate(components_stl_ts6):
    component.plot(ax=axes[i, 1], title=titles[1][i])
    axes[i, 1].grid()

# TS7 ETS components (Column 2)
components_ets_ts7 = [
    classified_ts_07["Value"],
    ets_classified_ts_07.trend,
    ets_classified_ts_07.seasonal,
    ets_classified_ts_07.resid
]
for i, component in enumerate(components_ets_ts7):
    component.plot(ax=axes[i, 2], title=titles[2][i])
    axes[i, 2].grid()

# TS7 STL components (Column 3)
components_stl_ts7 = [
    classified_ts_07["Value"],
    stl_classified_ts_07.trend,
    stl_classified_ts_07.seasonal,
    stl_classified_ts_07.resid
]
for i, component in enumerate(components_stl_ts7):
    component.plot(ax=axes[i, 3], title=titles[3][i])
    axes[i, 3].grid()

plt.suptitle("Comparative Graphs ETS vs. STL Decompositions", fontsize=24)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
