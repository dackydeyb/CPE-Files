# Pandas ( Lab Activity )




```python
import pandas as pd
```

```python
# Number 26
# Create the dataframe as variable df, using 20 random integer
np.random.seed(100)
df = pd.DataFrame(np.random.randint(0, 50, (5, 4)), ['R1', 'R2', 'R3', 'R4', 'R5'], ['A', 'B', 'C', 'D'])
df
```

```python
# Number 27
# What is the mean of column A
mean_Column_A = df['A'].mean()
mean_Column_A
```

```python
# Number 28
# What is the mean of column R4
mean_R4 = df.loc['R4'].mean()
mean_R4
```

```python
# Number 29
# What is the standard deviation of the column C and Column D?
std_dev = df[['C', 'D']].std()
std_dev
```

```python
# Number 30
# Create column E based from the sum of A and C
df['E'] = df['A'] + df['C']
df
```

```python
# Number 31
# Select only values from Column E that is greater that 60
result = df[df['E'] > 60][['A', 'C', 'E']]
result
```

```python
# Number 31
# Select only values from Column E that is greater than 60 and Column D that is greater than 25
result = df[(df['E'] > 60) & (df['D'] > 25)][['A', 'E', 'D']]
result
```

```python
# Number 32
# Using .iloc select the following subset of dataframe
subset = df.iloc[[1, 3, 4], 2:5]
subset
```

```python
# Number 33
# Using .iloc select the following subset of dataframe
subset = df.iloc[2:4, 3:]
subset
```

```python
# Number 34
# Drop colum E, and assign to a new varauble 'reduced_df'
reduced_df = df.drop('E', axis=1)
reduced_df
```
