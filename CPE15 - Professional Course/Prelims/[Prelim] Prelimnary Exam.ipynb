```python
import numpy as np
import pandas as pd
```

```python
# With a random seed of 80.
# Generate a 10 x 10 DataFrame with random integers from 0 to 100, assigned to variable 'df'
# Column names start from C1 to C10

np.random.seed(80)

df = pd.DataFrame(np.random.randint(0, 100, (10, 10)), columns=[f'C{i}' for i in range(1, 11)])

df
```

```python
# Question 1
# If you select the 5th row,
# with the 1st, 4th and 6th columns only.
# What will be the standard deviation of the whole dataframe
# In two decimals only, without rounding off.

selected_row = df.iloc[4, [0, 3, 5]]

std = df.std().std()

print(f"{std:.2f}")
```

```python
# Question No. 2
# If you select the 1st, 3rd, 4th, 8th row,
# with the 2nd, 4th, and 7th column only,
# What will be the the median of the 4th column?
# In two decimals only, without rounding off.

x = df.iloc[[0, 2, 3, 7], [1, 3, 6]]

y = x.iloc[:, 2].median()

print(f"{y:.2f}")
```

```python
# Question No. 3
# What is the 75th Percentile of the 7th Column
# In two decimals only, without rounding off.

percentile_75 = df['C7'].quantile(0.75)
print(f"{percentile_75:.2f}")
```

```python
# Question No. 4
# If you add row 2 and row 3, find the variance of the subset

x = df.iloc[[1, 2]].sum().var()
x
```

```python
# Question No. 5
# Select row 3 and row 4, column 5 to 7 then perform np.dot() to
# row 8 to row 10, column 8 to 10
# Write the resulting matrix

#Row 1 = index 0

x1 = df.iloc[2:4, 4:7]

x2 = df.iloc[7:10, 7:10]

y = np.dot(x1, x2)
y
```

```python
product = pd.read_csv('https://raw.githubusercontent.com/renatomaaliw3/public_files/refs/heads/master/Data%20Sets/products-y.csv')
product.head()
```

```python
# Question 06
# Which Store has the highest total sales in terms of Female Customers?

female_sales = product[product['Sex'] == 'Female'].groupby('Store')['Price'].sum()
x = female_sales.idxmax()
x
```

```python
# Question 07
#Average Bag Sales for Male at Store 5

x = product[(product['Item'] == 'Bag') & (product['Store'] == 'Store 5') & (product['Sex'] == 'Male')]['Price'].mean()

x
```

```python
# Question 08
# Which of the products are the most sold, neglect the pricing

most_sold = product.groupby('Item')['Price'].sum()
x = most_sold.idxmax()
x
```

```python
# Question 09
# Create another column from the original dataframe named 'Tax' which is 15% of the Price,
# then, find out which Store/s have paid the least taxes

product['Tax'] = product['Price'] * 0.15
x = product.groupby('Store')['Tax'].sum().idxmin()

x
```

```python
# Question 10
# What product is the least profitable for Women?

x = product[product['Sex'] == 'Female'].groupby('Item')['Price'].sum().idxmin()
x
```
