```python
# Load the following data

food = pd.read_csv('https://raw.githubusercontent.com/renatomaaliw3/public_files/refs/heads/master/Data%20Sets/foods.csv')
food.head()
```

```python
# 16. From the loaded dataset, using groupby, show the following condition
# Sum of Spend by 'Female'

food.groupby('Gender').sum()['Spend']
```

```python
# 17. From the loaded dataset, using groupby, show the following condition
# Average Spend by City, What City has the highest Average Spend?

food.groupby('City')['Spend'].mean()
```

```python
# 18. From the loaded dataset, using groupby, show the following condition
# Median Spend by Frequency, What City has the highest Median Spend?

food.groupby('City')['Spend'].median()
```

```python
# 19 From the loaded dateset, using combinations of .query, show the following condition
# You can hanve multiple lines of code to achieve this, not  necessarily one line
# Average Spend of Male from Philadelphia

food.query("City == 'Philadelphia'").groupby(by = ['Gender'])['Spend'].mean()

```

```python
# 20. From the loaded dataset, using combinations of .query or .groupby, show the following condition
# You can have multiple lines of code to achieve this, not necessarily one line
# Total Spend of Female on Sushi at New York
food.query("Item == 'Sushi' and  City == 'New York'").groupby(by = ['Gender'])['Spend'].sum()
```

```python
# 21. From the loaded dataset, using combinations of .query or .groupby, show the following condition
# You can have multiple lines of code to achieve this, not necessarily one line
# What is the Total Weekly Sales on Chalupa per City

food.query("Item == 'Chalupa' and Frequency == 'Weekly'").groupby(by = ['Item', 'City'])['Spend'].sum()


```
