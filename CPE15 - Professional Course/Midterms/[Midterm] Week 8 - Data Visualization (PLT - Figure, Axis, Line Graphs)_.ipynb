# Data Visualizations (Part - 01)

```python
# Imports

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```


# Introduction to 'plt.figure'

-- is a core function in Matplotlib, used to create and configure a new figure for plotting.

-- a "figure" in Matplotlib is an object that holds all elements of a plot, including axes, titles, labels, and other visual elements.


```python
# A. Figure (figure)

plt.figure()
plt.plot([1, 2, 3], [4, 5, 6]) # This is a basic line graph
plt.show()
```

```python

# B. (figsize)
# You can control the width and height of the figure in inches via figure size

plt.figure(figsize = (10, 3)) # 10 (width) 3 (height) tall
plt.plot([1, 2, 3], [4, 5, 6])
plt.show()
```

```python
# C. DPI (Dots Per Inch)

plt.figure(figsize = (10, 5), dpi = 100) # 100 DPI for moderate resolution
plt.plot([1, 2, 3], [4, 5, 6])
plt.show()
```

```python
# D. Background Color (facecolor)

plt.figure(figsize = (10,5), facecolor = '#abc123') # red background
plt.plot([1, 2, 3], [4, 5, 6])
plt.show()
```

```python
# E. Frame Visibility (frameon)

plt.figure(figsize = (10,5), facecolor = 'yellow', frameon = False) # Removes the border frame
plt.plot([1, 2, 3], [4, 5, 6])
plt.show()
```

```python
# F. Figure Title (frameon)
# While not a direct parameter of plt.figure, the title can be added to the figure
# using suptitle

fig = plt.figure()
fig.suptitle('Main Figure Title')
plt.plot([1, 2, 3], [4, 5, 6])
plt.show()
```

```python
# G. Saving a Figure (savefig)
# While not a direct parameter of plt.figure, the title can be added to the figure
# using suptitle

fig = plt.figure(figsize = (10, 5), facecolor = 'red', dpi = 100)
fig.suptitle('Main Figure Title')
plt.plot([1, 2, 3], [4, 5, 6])
fig.savefig('my_figure.png')
```

# Data Visualization Part 02 (Axes)


# A. Introduction to 'axes'

-- the area of a figure where data is plotted

-- it holds elements like **labels**, **titles**, **grids**, and **legends** and provides methods to manipulate them

-- they are crucial for creating and customizing plots, especially for multi-dimensional figures (Up to 3 dimensional only)


```python
# Axis Background

fig = plt.figure(figsize = (5, 5))

plt.gca().set_facecolor('#abe001') # Axis background

plt.plot([1, 2, 3], [4, 5, 6]) # This is a basic line graph
plt.show()
```

```python
# A. Axis Title

fig = plt.figure(figsize = (5, 5))
fig.suptitle('Main Figure Title') # This is a figure title, not the axes title

plt.title('Axes Title') # This is the axes title

plt.plot([1, 2, 3], [4, 5, 6]) # This is a basic line graph
plt.show()
```

```python
# B. Axis Labels

fig = plt.figure(figsize = (5, 5))
fig.suptitle('Main Figure Title') # This is a figure title, not the axes title
plt.title('Axes Title') # This is the axes title

plt.xlabel('X Axis Label') # X axis label
plt.ylabel('Y Axis Label') # Y axis label

plt.plot([1, 2, 3], [4, 5, 6]) # This is a basic line graph
plt.show()
```

```python
# C. Grids

fig = plt.figure(figsize = (5, 5))
fig.suptitle('Main Figure Title') # This is a figure title, not the axes title
plt.title('Axes Title') # This is the axes title
plt.xlabel('X Axis Label') # X axis label
plt.ylabel('Y Axis Label') # Y axis label
plt.plot([1, 2, 3], [4, 5, 6]) # This is a basic line graph

plt.grid(True, color = '#010abc', linestyle = 'dotted', linewidth = 1, alpha = 0.65)

plt.show()
```

```python
# D. Legends (loc, bbox_to_anchor)

fig = plt.figure(figsize = (5, 5))
fig.suptitle('Main Figure Title') # This is a figure title, not the axes title
plt.title('Axes Title') # This is the axes title
plt.xlabel('X Axis Label') # X axis label
plt.ylabel('Y Axis Label') # Y axis label

plt.plot([1, 2, 3], [4, 5, 6], label = 'Line 1') # This is a basic line graph
plt.plot([4, 5, 6], [1, 2, 3], label = 'Line 2') # This is a basic line graph
plt.grid(True, color = 'red', linestyle = 'dotted', linewidth = 1, alpha = 0.5)

plt.legend(loc = 'upper left', bbox_to_anchor = (0, 1))     # Adds Legend, e.g. loc = 'upper right', loc = 'center left', loc = 'center', loc = 'best'
                                                            # bbox_to_anchor = (x, y)
plt.show()
```

```python
# E. Ticks

fig = plt.figure(figsize = (7, 5))
fig.suptitle('Main Figure Title') # This is a figure title, not the axes title
plt.title('Axes Title') # This is the axes title
plt.xlabel('X Axis Label') # X axis label
plt.ylabel('Y Axis Label') # Y axis label

plt.plot([1, 2, 3], [4, 5, 6], label = 'Line 1') # This is a basic line graph
plt.plot([4, 5, 6], [1, 2, 3], label = 'Line 2') # This is a basic line graph

plt.grid(True, color = 'red', linestyle = 'dotted', linewidth = 1, alpha = 0.5)
plt.legend(loc = 'upper left', bbox_to_anchor = (0, 1))     # Adds Legend, e.g. loc = 'upper right', loc = 'center left', loc = 'center', loc = 'best'
                                                            # bbox_to_anchor = (x, y)

plt.xticks([1, 2, 3, 4, 5, 6, 7, 8]) # x ticks
plt.yticks([1, 2, 3, 4, 5, 6, 7, 8]) # y ticks

plt.show()
```

```python
# F. Limits

fig = plt.figure(figsize = (7, 5))
fig.suptitle('Main Figure Title') # This is a figure title, not the axes title
plt.title('Axes Title') # This is the axes title
plt.xlabel('X Axis Label') # X axis label
plt.ylabel('Y Axis Label') # Y axis label
plt.plot([1, 2, 3], [4, 5, 6], label = 'Line 1') # This is a basic line graph
plt.grid(True, color = 'red', linestyle = 'dotted', linewidth = 1, alpha = 0.5)
plt.legend(loc = 'upper left', bbox_to_anchor = (0, 1))     # Adds Legend, e.g. loc = 'upper right', loc = 'center left', loc = 'center', loc = 'best'
                                                            # bbox_to_anchor = (x, y)
plt.xticks(np.arange(1, 13, 2).tolist()) # x ticks
plt.yticks(np.arange(1, 13, 2).tolist()) # y ticks

plt.xlim(0, 13) # x limits
plt.ylim(0, 13) # y limits

plt.show()
```

```python
# Text Customizations

fig = plt.figure(figsize = (8, 8))


fig.suptitle('Main Figure Title', fontsize = 18, color = 'red', family = 'monospace', style = 'italic')
plt.title('Axes Title', fontdict = {'color': 'blue', 'fontsize': 14, 'fontweight': 'bold', 'family': 'sans-serif', 'style': 'italic'}) # This is the axes title


plt.xlabel('X Axis Label', fontdict = {'color': 'orange', 'fontsize': 12, 'fontweight': 'bold', 'family': 'sans-serif', 'style': 'italic'}) # X axis label
plt.ylabel('Y Axis Label', color = 'green', style = 'italic', family = 'monospace') # Y axis label


plt.plot([1, 2, 3], [4, 5, 6], label = 'Line 1') # This is a basic line graph
plt.grid(True, color = 'red', linestyle = 'dotted', linewidth = 1, alpha = 0.5)
plt.legend(loc = 'upper left', bbox_to_anchor = (0, 1))     # Adds Legend, e.g. loc = 'upper right', loc = 'center left', loc = 'center', loc = 'best'
                                                            # bbox_to_anchor = (x, y)
plt.xticks(np.arange(1, 10).tolist()) # x ticks
plt.yticks(np.arange(1, 10).tolist()) # y ticks

plt.tick_params(axis = 'y',                                   # both
                labelsize = 24,
                labelcolor = 'red',
                direction = 'inout',                          # in, out, inout
                length = 25,
                width = 4,
                pad = 30,
                labelrotation = -90)                         # angle

plt.tick_params(axis = 'x',
                labelsize = 14,
                labelcolor = 'red',
                direction = 'out',                          # in, out, inout
                length = 15,
                width = 2,
                pad = 15,
                labelrotation = 45)                         # angle


plt.xlim(0, 10) # x limits
plt.ylim(0, 10) # y limits

plt.show()
```


# B. Line Graphs, Customizing Line Styles

-- The general purpose of **line graphs** is to visually represent trends, changes, and patterns over time or across ordered data points.

-- **Showing Trends Over Time**: They effectively illustrate how a variable evolves over specific intervals, making them ideal for displaying time-series data, such as sales, temperature changes, or population growth. Each point represents a specific time, and connecting these points with a line provides a clear view of trends, such as increases, decreases, or consistency over time

-- **Comparing Multiple Variables**: Line graphs allow comparison of multiple datasets on the same graph, each with a separate line. This is useful in identifying relationships or disparities among different categories, like comparing stock prices of multiple companies over a period

-- **Highlighting Rate of Change**: Line graphs display not only the values but also the slopes of the lines between points, which indicate the rate at which changes occur. Steeper slopes signal faster changes, while flatter lines suggest stability

-- **Providing Clarity in Data Patterns**: They are straightforward and easy to interpret, which makes them a powerful tool for conveying complex data in a clear, concise way. They are suitable for audiences needing quick insights, such as in business or scientific presentations

-- **Spotting Peaks, Troughs, and Cycles**: Line graphs can highlight recurring patterns, like cycles or seasonal variations. They help identify anomalies, such as a sharp drop or unexpected peak, which can be critical for further analysis or decision-making


```python
# Color, Line Style, Line Width

fig = plt.figure(figsize = (5, 5))
fig.suptitle('Main Figure Title') # This is a figure title, not the axes title
plt.title('Axes Title') # This is the axes title
plt.xlabel('X Axis Label') # X axis label
plt.ylabel('Y Axis Label') # Y axis label
plt.grid(True)

plt.plot([1, 2, 3], [4, 5, 6],
         color = 'red',                 # color
         linestyle = 'dashed',          # linestyle (solid, dotted, dashed, dashdot, [parameterized])
         linewidth = 2)                 # any float values

plt.show()

# Link to more settings on linestyle : https://matplotlib.org/stable/gallery/lines_bars_and_markers
```

```python
# Horizontal and Vertical Lines

fig = plt.figure(figsize = (5, 5))
fig.suptitle('Main Figure Title') # This is a figure title, not the axes title
plt.title('Axes Title') # This is the axes title
plt.xlabel('X Axis Label') # X axis label
plt.ylabel('Y Axis Label') # Y axis label
plt.grid(True)

plt.plot([1, 2, 3], [4, 5, 6],
         color = 'red',             # color
         linestyle = 'dashed',      # linestyle (solid, dotted, dashed, dashdot, [parameterized])
         linewidth = 2)             # any float values

plt.axhline(y = 5.25, color = 'blue', linestyle = 'dotted', linewidth = 1)
plt.axvline(x = 2.15, color = 'green', linestyle = 'dashed', linewidth = 1)
plt.axvline(x = 2.75, color = 'green', linestyle = 'dashed', linewidth = 1)

plt.show()
```

```python
# Marker, Marker Size, Marker Edge Color, Mark Every

fig = plt.figure(figsize = (5, 5))
fig.suptitle('Main Figure Title') # This is a figure title, not the axes title
plt.title('Axes Title') # This is the axes title
plt.xlabel('X Axis Label') # X axis label
plt.ylabel('Y Axis Label') # Y axis label
plt.grid(True)

plt.plot([1, 2, 3], [4, 5, 6],
         color = 'red',                 # color
         linestyle = 'dashed',          # linestyle (solid, dotted, dashed, dashdot, [parameterized])
         linewidth = 2,                 # any float values
         marker = 'o',                  # sets the marker style
         markersize = 12,                # sets the marker size
         markeredgecolor = 'blue')      # sets the markeredgecolor

plt.show()

# Link to more settings on marker : https://matplotlib.org/stable/gallery/lines_bars_and_marker
```

```python
# Z Order, Pad for titles, Label Pad for axis, Loc for title, Sup Titles

fig = plt.figure(figsize = (8, 8))

fig.suptitle('Main Figure Title', fontsize = 18, x = '0.55', y = 1, color = "brown") # This is a figure title, not the axes title
plt.title('Axes Title', pad = 20, loc = 'left') # This is the axes title

plt.xlabel('X Axis Label', labelpad = 20) # X axis label
plt.ylabel('Y Axis Label', labelpad = 20) # Y axis label

plt.grid(True)

plt.plot([1, 2, 3], [4, 5, 6],
         label = 'Line 1',
         color = 'black',               # color
         linestyle = 'solid',           # linestyle (solid, dotted, dashed, dashdot, [parameterized])
         linewidth = 5,                 # any float values
         marker = 'o',                  # sets the marker style
         markersize = 12,               # sets the marker size
         markeredgecolor = 'blue',      # sets the markeredgecolor
         zorder = 11)                   # z-ordering

plt.plot([1, 2, 3], [8, 2, 3],
         label = 'Line 2',
         color = 'red',                 # color
         linestyle = 'solid',           # linestyle (solid, dotted, dashed, dashdot, [parameterized])
         linewidth = 5,                 # any float values
         marker = '*',                  # sets the marker style
         markersize = 25,               # sets the marker size
         markeredgecolor = 'black',     # sets the markeredgecolor
         zorder = 5)                    # z-ordering

plt.legend(loc = 'best')
plt.show()
```

```python
# Alpha

fig = plt.figure(figsize = (5, 5))
fig.suptitle('Main Figure Title') # This is a figure title, not the axes title
plt.title('Axes Title') # This is the axes title
plt.xlabel('X Axis Label') # X axis label
plt.ylabel('Y Axis Label') # Y axis label
plt.grid(True)

plt.plot([1, 2, 3], [4, 5, 6],
         label = 'Line 1',
         color = 'black',               # color
         linestyle = 'solid',           # linestyle (solid, dotted, dashed, dashdot, [parameterized])
         linewidth = 5,                 # any float values
         marker = 'o',                  # sets the marker style
         markersize = 12,               # sets the marker size
         markeredgecolor = 'blue',      # sets the markeredgecolor
         zorder = 10,                   # z-ordering
         alpha = 0.5)                   # transparency

plt.plot([1, 2, 3], [8, 2, 3],
         label = 'Line 2',
         color = 'red',                 # color
         linestyle = 'solid',           # linestyle (solid, dotted, dashed, dashdot, [parameterized])
         linewidth = 5,                 # any float values
         marker = '*',                  # sets the marker style
         markersize = 12,               # sets the marker size
         markeredgecolor = 'orange',    # sets the markeredgecolor
         zorder = 5)                    # z-ordering

plt.show()
```

```python
# Solid Capstyle, Dash Capstyle

fig = plt.figure(figsize = (5, 5))
fig.suptitle('Main Figure Title') # This is a figure title, not the axes title
plt.title('Axes Title') # This is the axes title
plt.xlabel('X Axis Label') # X axis label
plt.ylabel('Y Axis Label') # Y axis label
plt.grid(True)

plt.plot([1, 2, 3], [4, 5, 6],
         label = 'Line 1',                  # adds a label for the legend
         color = 'black',                   # color
         linestyle = 'solid',               # linestyle (solid, dotted, dashed, dashdot, [parameterized])
         linewidth = 15,                    # any float values
         #marker = 'o',                     # sets the marker style
         #markersize = 16,                  # sets the marker size
         markeredgecolor = 'blue',          # sets the markeredgecolor
         zorder = 10,                       # z-ordering
         alpha = 1,                         # transparency
         solid_capstyle = 'round')          # defines end shape of solid lines, 'butt', 'round', 'projecting'


plt.plot([1, 2, 3], [8, 2, 3],
         label = 'Line 2',
         color = 'red',                   # color
         linestyle = 'dashed',             # linestyle (solid, dotted, dashed, dashdot, [parameterized])
         linewidth = 15,                  # any float values
         #marker = '*',                   # sets the marker style
         #markersize = 12,                # sets the marker size
         markeredgecolor = 'orange',      # sets the markeredgecolor
         zorder = 5,
         dash_capstyle = 'projecting')          # defines end shape of dashed lines, 'butt', 'round', 'projecting'

plt.show()
```
